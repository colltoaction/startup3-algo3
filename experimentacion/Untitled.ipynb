```python
# -*- coding: utf-8 -*-
import pandas as pd
import sys
sys.path.append('/usr/local/lib/python2.7/dist-packages')
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats
import random
```

```python
!cd ../genetico && make experimentarFitness
```

```python
dfResultados = pd.read_csv('./fitness.out', delimiter=";")
dfVariance = dfResultados.groupby('Generacion').var()
dfMean = dfResultados.groupby('Generacion').mean()

ax = dfMean.plot(logy=False)
ax.set_ylabel("Media de la Fitness")
ax.set_xlabel("Generacion")
plt.title("Media de la Fitness en funcion de la generacion")
plt.show()


ax = dfVariance.plot(logy=False)
ax.set_ylabel("Varianza de la Fitness")
ax.set_xlabel("Generacion")
plt.title("Varianza de la Fitness en funcion de la generacion")
plt.show()
```
